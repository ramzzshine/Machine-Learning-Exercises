```python
X = data.loc[:,data.columns != 'price_doc'].values
y = data.loc[:,['price_doc']].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=None)
```

```python
models = []
models.append(("RFR", RandomForestRegressor()) )
models.append(("SVR", SVR()) )
#models.append(("XGR", XGBRegressor()) )
models.append(("MLP", MLPRegressor()) )
models.append(('GBR',GradientBoostingRegressor()))
scores = []
scores.append(("MAE", 'neg_mean_absolute_error'))
scores.append(("MSE", 'neg_mean_squared_error'))
scores.append(("RMSE", 'neg_root_mean_squared_error'))
scores.append(("R2", 'r2'))
```

```python
results = []
names = []
score_name =[]
for name, model in models:
    for score_n, score in scores:
        cv_results = model_selection.cross_val_score(model,X,y,cv=3, scoring=score)
        results.append(cv_results)
        names.append(name)
        score_name.append(score_n)
        msg = "{} : {} : {}".format(name,score_n , cv_results)
        print(msg)
```

```python
model = RandomForestRegressor(bootstrap=True,max_depth=80,max_features='auto',random_state=None,
                                  min_samples_leaf=2,min_samples_split=4,n_estimators=450,n_jobs = 1)
model.fit(X_train,y_train)
y_predict = model.predict(X_test)
r2score = r2_score(y_test,y_predict)

print(r2score)
print('Root Mean Squared Error:', sqrt(mean_squared_error(y_test, y_predict))) 
```

```python
def error(y_test,y_predict):
    print('mean absolute error(MAE):',mean_absolute_error(y_test, y_predict))
    print('mean squared error(MSE):',mean_squared_error(y_test, y_predict))
    print('root mean squared error(RMSE):',sqrt(mean_squared_error(y_test, y_predict)))
    print('R2:',r2_score(y_test, y_predict))
    print('explained variance score',explained_variance_score(y_test, y_predict))
```
